```python
# Import necessary Libary
from datasets import load_dataset # Import dataset from Huggingface Hub
import pandas as pd # Import for display data
```

```python
# Load Dataset from Huggingface Hub
dataset = load_dataset("bdstar/Tweets-Sentiment-Analysis")
```

```python
# Select dataset ranage for first 10 rows
df_display = pd.DataFrame(dataset['train'].select(range(10)))
```

```python
# Print the dataset
display(df_display)
```

```python
# Define the path to save the CSV file
csv_path = '/content/file.csv'
# Save the DataFrame to a CSV file
df_display.to_csv(csv_path, index=False)
# Print log for confirmation
print(f"Successfully saved the first {len(df_display)} rows to {csv_path}")
```
